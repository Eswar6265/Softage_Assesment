```python
from build123d import *

# --- Parameters ---

# Base Part Dimensions
bot_part_lower = 8.33     # Bottom Part Bottom Dimension
bot_part_mid = 9.33       # Bottom Part Middle Dimension
bot_part_upper = 9.33     # Bottom Part Upper Dimension
h_bot_part_1 = 0.5        # Bottom Part base_1 height         
h_bot_part_2 = 1.118      # Bottom Part base_2 height             
tot_ht_bot_part = 1.818   # Bottom Part base_1 + base_2 + 0.2 height         

# SHARED DEPTH
depth = 5.03              # Width of Bottom Part          

# Top Part Dimensions
top_part_lower_dim = depth   # Top Part lower dimensions
top_part_upper_dim = 3.223   # Top Part upper dimensions
ht_top_part = 1.992           # Height of Top Part (total height of bottom part + top part)
top_box_ext_length = 2.79    # Length of Top Part

# Extrusion Dimensions
extrusion_width = 2.589
extrusion_height = 1.992
extrusion_length = 1.999

# Top Attachment Part Dimensions
attachment_width = 1.49       # Top Attachment Width dimensions
attachment_height = 1.992     # Top Attachment Height dimensions
attachment_length = 0.746     # Top Attachment Length dimensions

# Mid Left & Right Part Dimensions
mid_attachment_width = 2.59       # Mid Attachment Width dimensions
mid_attachment_height = 1.992     # Mid Attachment Height dimensions
mid_attachment_length = 1.5       # Mid Attachment Length dimensions

# Mid Part Dimensions
mid_width = 2.39       # Mid Width dimensions
mid_height = 1.793     # Mid Height dimensions
mid_length = 3.256     # Mid Length dimensions

# Bottom Attachment Dimensions
bot_attachment_width_1 = 2
bot_attachment_width_2 = 2.8
bot_attachment_length_1 = 0.913
bot_attachment_length_2 = 3.074 
bot_attachment_length_3 = 3.6
bot_attachment_height = 0.2

# Side Attachment Dimensions
side_attachment_width = 2
side_attachment_length = 0.2
side_attachment_height = 2.237

# Side Extrusion Dimensions
side_extrusion_width = 0.837
side_extrusion_length = 0.2
side_extrusion_height = 1.028
side_extrusion_base = 0.605

# Side Protrusion Dimensions
side_protrusion_width = 0.8
side_protrusion_length = 0.2
side_protrusion_height = 0.884
side_protrusion_base = 0.7

# Small Part Dimensions
small_part_base_width = 2.589/2
small_part_btm_pnt_1_width = 0.55
small_part_btm_pnt_2_width = 0.625
small_part_top_pnt_2_width = 0.1
small_part_btm_pnt_1_height = 0.484
small_part_btm_pnt_2_height = 0.555
small_part_top_pnt_1_height = 1.078
small_part_top_pnt_2_height = 1.555
small_part_length = 0.749

# Top Part Planes
left_plane = Plane(
    origin=(-bot_part_upper/2, 0, tot_ht_bot_part - 0.130000248),
    x_dir=Plane.YZ.x_dir,
    z_dir=Plane.YZ.z_dir
)  # Left End Plane

right_plane = Plane(
    origin=(bot_part_upper/2, 0, tot_ht_bot_part - 0.130000248),
    x_dir=Plane.YZ.x_dir,
    z_dir=Plane.YZ.z_dir
) # Right End Plane 

# Extrusion Planes
Extrusion_left_plane = left_plane.offset(0.791) # Left Extrusion Plane Offset

Extrusion_right_plane = right_plane.offset(-0.791) # Left Extrusion Plane Offset

# Mid Part Planes

Mid_left_plane = left_plane.offset(1.537) # Left Extrusion Plane Offset

Mid_right_plane = right_plane.offset(-1.537) # Left Extrusion Plane Offset

Mid_part_plane = left_plane.offset(3.037) # Left Extrusion Plane Offset

small_part_plane = left_plane.offset(2.79)

with BuildPart() as Final_Part:

    # 1. Build the Base Block
    with BuildPart() as Base_Block:
    
        with BuildSketch(Plane.XZ):
            Polygon(
                (-bot_part_lower / 2, 0.2),         
                (bot_part_lower / 2, 0.2),          
                (bot_part_mid / 2, (h_bot_part_1+0.2)),      
                (bot_part_upper / 2, (tot_ht_bot_part)),    
                (-bot_part_upper / 2, (tot_ht_bot_part)),   
                (-bot_part_mid / 2, (h_bot_part_1+0.2))      
            )
        extrude(amount=depth / 2, both=True)

    add(Base_Block.part)

    # 2. Build the Top Left Block
    with BuildPart() as Top_Left_Block:
        
        with BuildSketch(left_plane):
            Polygon(
                (-depth / 2, 0),         
                (depth / 2, 0),             
                (top_part_upper_dim / 2, ht_top_part),   
                (-top_part_upper_dim / 2, ht_top_part)      
            )
        extrude(amount=top_box_ext_length , both=False)

        outer_face = Top_Left_Block.faces().sort_by(Axis.Z)[-1]
        outer_edges = list(outer_face.edges())

        fillet(outer_edges, radius=0.3)

    add(Top_Left_Block.part)

    # 3. Build the Top Right Block
    with BuildPart() as Top_Right_Block:
        
        with BuildSketch(right_plane):
            Polygon(
                (-depth / 2, 0),         
                (depth / 2, 0),             
                (top_part_upper_dim / 2, ht_top_part),   
                (-top_part_upper_dim / 2, ht_top_part)      
            )
        extrude(amount=-top_box_ext_length , both=False)

        outer_face = Top_Right_Block.faces().sort_by(Axis.Z)[-1]
        outer_edges = list(outer_face.edges())

        fillet(outer_edges, radius=0.3)
  
    add(Top_Right_Block.part)
    
    # 4. Modify the Top Left Block Cut
    with BuildPart() as Top_Left_Cut:
        
        with BuildSketch(Extrusion_left_plane):
            Polygon(
                (-extrusion_width / 2, 0),         
                (extrusion_width / 2, 0),             
                (extrusion_width / 2, extrusion_height),   
                (-extrusion_width / 2, extrusion_height)      
            )
        extrude(amount=extrusion_length , both=False)

    Final_Part.part -= Top_Left_Cut.part
    
    # 5. Modify the Top Right Block Cut
    with BuildPart() as Top_Right_Cut:
        
        with BuildSketch(Extrusion_right_plane):
            Polygon(
                (-extrusion_width / 2, 0),         
                (extrusion_width / 2, 0),             
                (extrusion_width / 2, extrusion_height),   
                (-extrusion_width / 2, extrusion_height)      
            )
        extrude(amount=-extrusion_length , both=False)

    Final_Part.part -= Top_Right_Cut.part
    
    # 6. Final Top Left Block
    with BuildPart() as Final_Top_Left_Block:
        
        with BuildSketch(Extrusion_left_plane):
            Polygon(
                (-attchment_width / 2, 0),         
                (attchment_width / 2, 0),             
                (attchment_width / 2, attchment_height),   
                (-attchment_width / 2, attchment_height)      
            )
        extrude(amount=attchment_length , both=False)

        top_face = Final_Top_Left_Block.faces().sort_by(Axis.X)[-1]
        top_edges = top_face.edges()
        top_inner_edge = max(top_edges, key=lambda e: e.center().Z)
        
        chamfer(top_inner_edge, length=0.285)

    add(Final_Top_Left_Block.part)
    
    # 7. Final Top Right Block
    with BuildPart() as Final_Top_Right_Block:
        
        with BuildSketch(Extrusion_right_plane):
            Polygon(
                (-attchment_width / 2, 0),         
                (attchment_width / 2, 0),             
                (attchment_width / 2, attchment_height),   
                (-attchment_width / 2, attchment_height)      
            )
        extrude(amount=-attchment_length , both=False)

        top_face = Final_Top_Right_Block.faces().sort_by(Axis.X)[0]
        top_edges = top_face.edges()
        top_inner_edge = max(top_edges, key=lambda e: e.center().Z)
        
        chamfer(top_inner_edge, length=0.285)

    add(Final_Top_Right_Block.part)
    
    # 8. Mid Left Block
    with BuildPart() as Mid_Left_Block:
        
        with BuildSketch(Mid_left_plane):
            Polygon(
                (-mid_attchment_width / 2, 0),         
                (mid_attchment_width / 2, 0),             
                (mid_attchment_width / 2, mid_attchment_height),   
                (-mid_attchment_width / 2, mid_attchment_height)      
            )
        extrude(amount=mid_attchment_length , both=False)

        outer_face = Mid_Left_Block.faces().sort_by(Axis.X)[0]

        x_edges = list(Mid_Left_Block.edges().filter_by(Axis.X))
        outer_edges = list(outer_face.edges())

        target_edges = list(set(x_edges + outer_edges))

        fillet(target_edges, radius=0.3)

    add(Mid_Left_Block.part)
    
    # 9.  Mid Right Block
    with BuildPart() as Mid_Right_Block:
        
        with BuildSketch(Mid_right_plane):
            Polygon(
                (-mid_attchment_width / 2, 0),         
                (mid_attchment_width / 2, 0),             
                (mid_attchment_width / 2, mid_attchment_height),   
                (-mid_attchment_width / 2, mid_attchment_height)      
            )
        extrude(amount=-mid_attchment_length , both=False)

        outer_face = Mid_Right_Block.faces().sort_by(Axis.X)[-1]

        x_edges = list(Mid_Right_Block.edges().filter_by(Axis.X))
        outer_edges = list(outer_face.edges())

        target_edges = list(set(x_edges + outer_edges))

        fillet(target_edges, radius=0.3)

    add(Mid_Right_Block.part)
    
    # 10.  Mid Block
    with BuildPart() as Mid_Block:
        
        with BuildSketch(Mid_part_plane):
            Polygon(
                (-mid_width / 2, 0.1),         
                (mid_width / 2, 0.1),             
                (mid_width / 2, mid_height + 0.1),   
                (-mid_width / 2, mid_height + 0.1)      
            )
        extrude(amount= mid_length , both=False)

        fillet(
        Mid_Block.edges().filter_by(Axis.X),
        radius=0.2
    )

    add(Mid_Block.part)

    # 11. Bottom Attachment Left Part
    with BuildPart() as Bottom_Attachment_Left:

        bottom_z = bottom_face.center().Z # Assuming bottom_face is defined earlier in your script

        Bottom_Plane = Plane(
            origin=(0, 0, bottom_z),
            x_dir=(1, 0, 0),
            z_dir=(0, 0, 1)
        )

        with BuildSketch(Bottom_Plane):
            with BuildLine():
                Polyline(
                    (-bot_part_upper / 2, bot_attachment_width_1/2),
                    (-(bot_part_upper / 2) + bot_attachment_length_1, bot_attachment_width_1/2 ),
                    (-(bot_part_upper / 2) + bot_attachment_length_1, bot_attachment_width_2/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_2, bot_attachment_width_2/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_2, bot_attachment_width_1/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_3, bot_attachment_width_1/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_3, -bot_attachment_width_1/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_2, -bot_attachment_width_1/2 ),
                    (-(bot_part_upper / 2) +  bot_attachment_length_2, -bot_attachment_width_2/2 ),
                    (-(bot_part_upper / 2) + bot_attachment_length_1, -bot_attachment_width_2/2 ),
                    (-(bot_part_upper / 2) + bot_attachment_length_1, -bot_attachment_width_1/2 ),
                    (-bot_part_upper / 2, -bot_attachment_width_1/2),
                    close=True
                )
            make_face()
            
        extrude(amount= -bot_attachment_height , both=False)

    add(Bottom_Attachment_Left.part)

    # 12. Bottom Attachment Right Part
    with BuildPart() as Bottom_Attachment_Right:
    
        with BuildSketch(Bottom_Plane):
            with BuildLine():
                Polyline(
                    (bot_part_upper / 2, bot_attachment_width_1/2),
                    ((bot_part_upper / 2) - bot_attachment_length_1, bot_attachment_width_1/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_1, bot_attachment_width_2/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_2, bot_attachment_width_2/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_2, bot_attachment_width_1/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_3, bot_attachment_width_1/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_3, -bot_attachment_width_1/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_2, -bot_attachment_width_1/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_2, -bot_attachment_width_2/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_1, -bot_attachment_width_2/2 ),
                    ((bot_part_upper / 2) - bot_attachment_length_1, -bot_attachment_width_1/2 ),
                    (bot_part_upper / 2, -bot_attachment_width_1/2),
                    close=True
                )
            make_face()
            
        extrude(amount= -bot_attachment_height , both=False)

    add(Bottom_Attachment_Right.part)

    # 13. Side Attachment Left Part
    
    plane_z = left_plane.origin.Z
    target_z = Bottom_Attachment_Right.part.bounding_box().min.Z 
    y_start = target_z - plane_z
    
    with BuildPart() as Side_Attachment_Left:
        
        with BuildSketch(left_plane):
            with BuildLine():
                Polyline(
                (-side_attachment_width / 2, y_start),         
                (side_attachment_width / 2, y_start),             
                (side_attachment_width / 2, side_attachment_height + y_start),   
                (-side_attachment_width / 2, side_attachment_height + y_start),
                close=True
            )
            make_face()
            
        extrude(amount= -side_attachment_length , both=False)

        top_edges_left = Side_Attachment_Left.edges().filter_by(Axis.X).sort_by(Axis.Z)[-2:]
        fillet(top_edges_left, radius=0.5)

    add(Side_Attachment_Left.part)
    
    # 14. Side Attachment Right Part
    with BuildPart() as Side_Attachment_Right:
        
        with BuildSketch(right_plane):
            with BuildLine():
                Polyline(
                (-side_attachment_width / 2, y_start),         
                (side_attachment_width / 2, y_start),             
                (side_attachment_width / 2, side_attachment_height + y_start),   
                (-side_attachment_width / 2, side_attachment_height + y_start),
                close=True
            )
            make_face()
            
        extrude(amount= side_attachment_length , both=False)

        top_edges_right = Side_Attachment_Right.edges().filter_by(Axis.X).sort_by(Axis.Z)[-2:]
        fillet(top_edges_right, radius=0.5)

    add(Side_Attachment_Right.part)

    # 15. Side Extrusion Left 
    
    with BuildPart() as Side_Extrusion_Left:
        
        with BuildSketch(left_plane):
            with BuildLine():
                Polyline(
                (-side_extrusion_width / 2, y_start + side_extrusion_base),         
                (side_extrusion_width / 2, y_start + side_extrusion_base),             
                (side_extrusion_width / 2, side_extrusion_height + y_start + side_extrusion_base),   
                (-side_extrusion_width / 2, side_extrusion_height + y_start + side_extrusion_base),
                close=True
            )
            make_face()
            
        extrude(amount= -side_extrusion_length , both=False)

    Final_Part.part -= Side_Extrusion_Left.part
    
    # 16. Side Attachment Right Part
    with BuildPart() as Side_Attachment_Right:
        
        with BuildSketch(right_plane):
            with BuildLine():
                Polyline(
                (-side_extrusion_width / 2, y_start + side_extrusion_base),         
                (side_extrusion_width / 2, y_start + side_extrusion_base),             
                (side_extrusion_width / 2, side_extrusion_height + y_start + side_extrusion_base),   
                (-side_extrusion_width / 2, side_extrusion_height + y_start + side_extrusion_base),
                close=True
            )
            make_face()
            
        extrude(amount= side_extrusion_length , both=False)

    Final_Part.part -= Side_Attachment_Right.part

    # 17. Side Protrusion Left 
    
    with BuildPart() as Side_Protrusion_Left:
        
        with BuildSketch(left_plane):
            with BuildLine():
                Polyline(
                (-side_protrusion_width / 2, y_start + side_protrusion_base),         
                (side_protrusion_width / 2, y_start + side_protrusion_base),             
                (side_protrusion_width / 2, side_protrusion_height + y_start + side_protrusion_base),   
                (-side_protrusion_width / 2, side_protrusion_height + y_start + side_protrusion_base),
                close=True
            )
            make_face()
            
        extrude(amount= -side_protrusion_length , both=False)

        outer_face = Side_Protrusion_Left.faces().sort_by(Axis.X)[0]
        horizontal_edges = outer_face.edges().filter_by(Axis.Y)
        bottom_outer_edge = horizontal_edges.sort_by(Axis.Z)[0]
        chamfer(bottom_outer_edge, length=0.199)

    add(Side_Protrusion_Left.part)
    
    # 18. Side Protrusion Right 
    with BuildPart() as Side_Protrusion_Right:
        
        with BuildSketch(right_plane):
            with BuildLine():
                Polyline(
                (-side_protrusion_width / 2, y_start + side_protrusion_base),         
                (side_protrusion_width / 2, y_start + side_protrusion_base),             
                (side_protrusion_width / 2, side_protrusion_height + y_start + side_protrusion_base),   
                (-side_protrusion_width / 2, side_protrusion_height + y_start + side_protrusion_base),
                close=True
            )
            make_face()
            
        extrude(amount= side_protrusion_length , both=False)

        outer_face = Side_Protrusion_Right.faces().sort_by(Axis.X)[-1]
        horizontal_edges = outer_face.edges().filter_by(Axis.Y)
        bottom_outer_edge = horizontal_edges.sort_by(Axis.Z)[0]
        chamfer(bottom_outer_edge, length=0.199)

    add(Side_Protrusion_Right.part)

    # 20. Final Bottom Plate Fillets
    
    bottom_faces = Final_Part.faces().sort_by(Axis.Z)[:2]
    y_edges = bottom_faces.edges().filter_by(Axis.Y)
    sorted_y_edges = y_edges.sort_by(Axis.X)
    target_edges = [sorted_y_edges[0], sorted_y_edges[-1]]
    fillet(target_edges, radius=0.19)
    
# --- Export ---
export_stl(Final_Part.part, "Final_Part.stl")
```
